```python
import pandas as pd
from sqlalchemy import create_engine

# --------------------------------
# CONFIG
# --------------------------------
CSV_FILE = "../yellow_tripdata_2024-01.csv"
TABLE_NAME = "yellow_taxi_data"
```

```python
# --------------------------------
# 1) Lectura completa del CSV
# --------------------------------
df = pd.read_csv(CSV_FILE)
```

```python
# --------------------------------
# 2) Conexión a Postgres
# --------------------------------
engine = create_engine(
    "postgresql://root:root@localhost:5432/ny_taxi"
)
```

```python
# --------------------------------
# 3) Creación de tabla (solo esquema)
# --------------------------------
df.head(n=0).to_sql(
    name=TABLE_NAME,
    con=engine,
    if_exists="replace"
)
```

```python
# --------------------------------
# 4) Carga completa de datos
# --------------------------------
%time df.to_sql(name=TABLE_NAME, con=engine, if_exists="append")
```
